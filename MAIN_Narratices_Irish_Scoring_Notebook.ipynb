```python
# =============================================================================
# MAIN Irish Assessment — Google Colab Batch Scorer

# =============================================================================


# ── Cell 1: Mount Google Drive and set up paths ───────────────────────────────

from google.colab import drive
drive.mount('/content/drive')

import sys, os

PROJECT_PATH = '/content/drive/MyDrive/MAIN_Irish'

sys.path.insert(0, PROJECT_PATH)
os.chdir(PROJECT_PATH)

print("✓ Drive mounted")
print(f"  Working directory: {os.getcwd()}")


```

```python
# ── Cell 2: Import the scorer ─────────────────────────────────────────────────

# The scorer file must be in PROJECT_PATH (see folder structure above)
from main_irish_scorer import process_transcripts, load_transcript

print("✓ MAIN Irish Scorer loaded")
```

```python
# ── Cell 3: Batch-process all transcripts ─────────────────────────────────────
#
# process_transcripts() will:
#   • Find every .txt file in the transcripts/ folder
#   • Skip any file whose name doesn't contain dog / cat / birds / goats
#   • Extract only the child's speech (SP02 turns)
#   • Score Sections A, B, C and D
#   • Print a full report for each transcript
#   • Save results to a CSV
#
# Optional arguments:
#   story_filter = "dog"    → only process dog transcripts
#   child_speaker = "SP02"  → override auto-detection (usually not needed)
#   output_csv = "..."      → change the output filename

TRANSCRIPTS_FOLDER = os.path.join(PROJECT_PATH, 'transcripts')
OUTPUT_CSV         = os.path.join(PROJECT_PATH, 'main_scores.csv')

process_transcripts(
    transcripts_dir = TRANSCRIPTS_FOLDER,
    output_csv      = OUTPUT_CSV,
    # story_filter  = "dog",      # uncomment to process one story at a time
    # child_speaker = "SP02",     # uncomment to override speaker detection
)

print(f"\n✓ Results saved to: {OUTPUT_CSV}")
```

```python
# ── Cell 4: View results as a table ──────────────────────────────────────────

import pandas as pd

df = pd.read_csv(OUTPUT_CSV)
print(f"Loaded {len(df)} scored transcript(s)\n")

# Core columns — edit this list to show what you need
display_cols = [
    'transcript_id', 'story', 'method',
    'A_Story_Total',
    'B1_AO', 'B2_G', 'B3_GA_GO', 'B4_GAO',
    'C_Total_IST',
    'D_Total',
]

# Only show columns that actually exist (D_Total absent if no comp questions)
display_cols = [c for c in display_cols if c in df.columns]

print(df[display_cols].to_string(index=False))
```

```python
# ── Cell 5: Summary statistics ────────────────────────────────────────────────

numeric_cols = [c for c in display_cols if c not in ('transcript_id', 'story', 'method')]

print("Summary statistics:\n")
print(df[numeric_cols].describe().round(2).to_string())
```

```python
# ── Cell 6: Export to Excel (optional) ───────────────────────────────────────

OUTPUT_XLSX = os.path.join(PROJECT_PATH, 'main_scores.xlsx')

df.to_excel(OUTPUT_XLSX, index=False, engine='openpyxl')
print(f"✓ Excel file saved to: {OUTPUT_XLSX}")
```
